```python
import datetime as dt
import sqlite3


my_conn = sqlite3.connect('kambalpizza.db')

from tkinter import messagebox
import tkinter as tk 
from tkinter import * 
my_w = tk.Tk()
my_w.geometry("1100x600") 
my_w.title("KAMBAL PIZZA!!")

#bg= PhotoImage(file="pizza/axel.png")
my_label=Label(my_w,bg="yellow")
my_label.place(x=0,y=0,relwidth=1,relheight=1)

label0 = tk.Label(my_w,  text='WELCOME KAMBAL PIZZA',font="Arial 25 bold italic", bg="yellow")            
label0.place(x=300,y=10)
prices = [[60,110,150,200],
         [50,120,180,220],
         [75,150,170,350],
         [80,130,170,340],
         [60,110,150,300],
         [75,135,150,321],
         [95,150,210,300]]
total_cost = 0
def calculate():
    global total_cost
    pizza, size =-1,-1
    if radio_v.get() == 'CHEESY ALOHA':pizza = 0
    elif radio_v.get() == 'CHEESY PEPERONI':pizza =1
    elif radio_v.get() == 'CHEESY MEAT':pizza =2
    elif radio_v.get() == 'CHEESY LOADED':pizza =3
    elif radio_v.get() == 'CHEESY HAM':pizza =4
    elif radio_v.get() == 'CHEESY VEGIES':pizza = 5
    elif radio_v.get() == 'CHEESY BURGER':pizza = 6
    
    if radio_x.get() == 'SMALL': size = 0
    elif radio_x.get() == 'MEDIUM': size =1
    elif radio_x.get() == 'LARGE': size =2
    elif radio_x.get() == 'EXTRA LARGE': size = 3
    
    total = prices[pizza][size]
    total_cost =(total)* int(t3.get())
    
    
    total1 = Label(my_w, text=total_cost,font ="Arial 15  italic") 
    total1.place(x=280,y=500)

#INFORMATION 
label1 = tk.Label(my_w,  text='Full Name: ', width=10, bg="yellow")  
label1.place(x=20,y=60)
t1 = tk.Text(my_w,  height=1, width=25,bg='white') 
t1.place(x=150,y=60)



labelx1 = tk.Label(my_w,  text='Phone Number: ', width=10,bg="yellow")  
labelx1.place(x=20,y=90)
x1 = tk.Text(my_w,  height=1, width=25,bg='white') 
x1.place(x=150,y=90)

labelx2 = tk.Label(my_w,  text='Address: ', width=10,bg="yellow")  
labelx2.place(x=20,y=120)
x2 = tk.Text(my_w,  height=1, width=25,bg='white') 
x2.place(x=150,y=120)

#Date
now = dt.datetime.now()
timestr1= now.strftime("%H:%M:%S:%p")
labelx3 =tk.Label(my_w,  text=timestr1,font="Arial 15 bold italic", width=50,bg="yellow")  
labelx3.place(x=350,y=60)
date = tk.Label(my_w,  text="TIME:", width=10,bg="yellow")  
date.place(x=430,y=60)


#Time
datestr= now.strftime("%Y-%m-%d")
labelx4 = tk.Label(my_w,  text=datestr,font="Arial 15 bold italic", width=50,bg="yellow")  
labelx4.place(x=350,y=100)
time = tk.Label(my_w,  text="DATE:" , width=10,bg="yellow")  
time.place(x=430,y=100)




#SIZE OF PIZZA
label2 = tk.Label(my_w,  text='Payment Method',font="Arial 15 bold italic", width=15,bg="yellow")  
label2.place(x=390,y=170) 
options = StringVar(my_w)
options.set("") 

opt1 = OptionMenu(my_w, options, "COD", "Credit", "Cash in")
opt1.place(x=370,y=190) 

#Quantity
label3 = tk.Label(my_w,  text='Quantity: ', font="Arial 20 bold italic",width=10,bg="yellow")  
label3.place(x=370,y=220) 


t3 =tk.Entry(my_w,width = 10)
t3.insert(0, "0")
t3.place(x=380,y=255)



# PIZZA FLAVOR 
label4 = Label(my_w, text=" CHOOSE YOUR PIZZA ",font="Arial 20 bold italic",bg="yellow") 
label4.place(x=50,y=160)

radio_v = tk.StringVar()
radio_v.set("ALOHA")

radio_t = tk.StringVar()
radio_t.set('PIZZA')
name = "CHEESY ALOHA"
name1 = "CHEESY PEPERONI"
name2 = "CHEESY MEAT"
name3 = "CHEESY LOADED"
name4 = "CHEESY HAM"
name5 = "CHEESY VEGIES"
name6 = "CHEESY BURGER"
def clicked(value):
    myLabel = Label (my_w, text = value)
    myLabel.place(x =500,y = 500)


rad= Radiobutton(my_w, text =name, font="Arial 17 ", bg="yellow",variable=radio_v,  value = name, command = lambda: clicked(())).place(x =50,y = 200)
rad2=Radiobutton(my_w, text =name1, font="Arial 17 ", bg="yellow",variable=radio_v, value = name1, command = lambda: clicked(())).place(x =50,y = 240)
rad3=Radiobutton(my_w, text =name2, font="Arial 17 ", bg="yellow",variable=radio_v, value = name2, command = lambda: clicked(())).place(x =50,y = 280)
rad3=Radiobutton(my_w, text =name2, font="Arial 17 ", bg="yellow",variable=radio_v, value = name2, command = lambda: clicked(())).place(x =50,y = 280)
rad4=Radiobutton(my_w, text =name3, font="Arial 17 ", bg="yellow",variable=radio_v, value = name3, command = lambda: clicked(())).place(x =50,y = 320)
rad5=Radiobutton(my_w, text =name4, font="Arial 17 ", bg="yellow",variable=radio_v, value = name4, command = lambda: clicked(())).place(x =50,y = 360)
rad6=Radiobutton(my_w, text =name5, font="Arial 17 ", bg="yellow",variable=radio_v, value = name5, command = lambda: clicked(())).place(x =50,y = 400)
rad7=Radiobutton(my_w, text =name6, font="Arial 17 ", bg="yellow",variable=radio_v, value = name6, command = lambda: clicked(())).place(x =50,y = 440)



#CHECK OUT BUTTON
#sum
totaltext = Label(my_w, text=" Total Bill:\t\t â‚± ",font ="Arial 15  italic",bg="yellow") 
totaltext.place(x=100,y=500)

          
bx2= Button(my_w , text = "Add Product",width =20 ,command = calculate ,bg="yellow")
bx2.place(x=205,y=520)




button1 = tk.Button(my_w,  text='CHECK OUT',width=15, 
               command=lambda: add_data())
button1.place(x=800,y=450)


my_str = tk.StringVar()
label5 = tk.Label(my_w,  textvariable=my_str, width=10,bg="yellow" )  
label5.place(x=800,y=500)
my_str.set(" ")

#MENU 
labe28 = Label(my_w, text=" MENU",font ="Arial 20 bold italic",bg="yellow") 
labe28.place(x=760,y=150)
labe27 = Label(my_w, text=" \t\tSMALL \t MEDIUM  LARGE\t EXTRA LARGE ",font ="Arial 15 bold italic",bg="yellow") 
labe27.place(x=570,y=180)
label6 = Label(my_w, text=" Cheesy Aloha\t60php\t 110php\t 150php\t 200php",font ="Arial 15  italic",bg="yellow") 
label6.place(x=570,y=210)
label7 = Label(my_w, text=" Cheesy Peperoni\t50php\t 120php\t 180Php\t 220php",font ="Arial 15  italic",bg="yellow") 
label7.place(x=570,y=240)
label8 = Label(my_w, text=" Cheesy Meat\t75php\t 150php\t 170php\t 350php",font ="Arial 15  italic",bg="yellow") 
label8.place(x=570,y=270)
label9 = Label(my_w, text=" Cheesy Loaded\t80PHP\t 130php\t 170php\t 340php",font ="Arial 15  italic",bg="yellow") 
label9.place(x=570,y=300)
labe20 = Label(my_w, text=" Cheesy Ham \t60PHP\t 110php\t 150php\t 300php",font ="Arial 15  italic",bg="yellow") 
labe20.place(x=570,y=330)
labe21 = Label(my_w, text=" Cheesy Vegies \t75PHP\t 135php\t 150php\t 321php",font ="Arial 15  italic",bg="yellow") 
labe21.place(x=570,y=360)
labe23 = Label(my_w, text=" Cheesy Burger \t95PHP\t 150php\t 210php\t 300php",font ="Arial 15  italic",bg="yellow") 
labe23.place(x=570,y=390)


#MODE OF PAYMENT
label25 = Label(my_w, text=" Size ",font="Arial 15 bold italic",bg="yellow") 
label25.place(x=370,y=280)

radio_x = tk.StringVar()
radio_x.set('x')
rtx1 = tk.Radiobutton(my_w, text='Small ',font="Arial 17 ", bg="yellow", variable=radio_x, value='SMALL')
rtx1.place(x=370,y=310)

rtx2 = tk.Radiobutton(my_w, text='Medium ',font="Arial 17 ", bg="yellow", variable=radio_x, value='MEDIUM')
rtx2.place(x=370,y=340)

rtx3 = tk.Radiobutton(my_w, text='Large ',font="Arial 17 ", bg="yellow", variable=radio_x, value='LARGE')
rtx3.place(x=370,y=370)

rtx4 = tk.Radiobutton(my_w, text='Extra Large ',font="Arial 17 ", bg="yellow", variable=radio_x, value='EXTRA LARGE')
rtx4.place(x=370,y=400) 




def closed():
    quit()



def add_data():
     res = messagebox.askquestion('askquestion', 'Do you want to Proceed?')
     if res == 'yes':
         messagebox.showinfo('Response', 'Added Data')
     elif res == 'no':
         messagebox.showinfo('Response', 'Cancelled Data.')
         return closed()

         

     flag_validation=True # set the flag 
     NAME=t1.get("1.0",END) # read name
     my_phone=x1.get("1.0",END) # read name
     my_address=x2.get("1.0",END) # read name
     my_DOR=str(datestr) # read name
     my_time=str(timestr1) # read name
     my_class=radio_x.get()    # read class
     my_age=t3.get() # read age
     my_pizza=radio_v.get() 
     my_gender=str(total_cost)  # read gender 
     my_MOP=options.get()   # read gender
     
  
        

   
   
        
    
     
     # length of my_name , my_class and my_gender more than 2 
     if(len(NAME) < 2 or len  (my_phone) < 2 or len  (my_address) < 2 or len  (my_DOR) < 2 or len  (my_time) < 2 or len(my_class) <2 or len(my_pizza) <2 or len(my_gender) < 2 or len(my_MOP) < 2 ):
            flag_validation=False 

     if(flag_validation):
        my_str.set("Adding data...")
        try:
            
            print("Connected to database successfully")
            
            
            my_conn.execute ("CREATE TABLE IF NOT EXISTS KAMBALPIZZA(NAME TEXT ,PHONE TEXT, ADDRESS TEXT, DATE_OF_ORDER TEXT,TIME TEXT,SIZE INT,QUANTITY INT , PRICE  TEXT,FLAVOR TEXT,MODE_OF_PAYMENT TEXT)")
            my_query="INSERT INTO KAMBALPIZZA values(?,?,?,?,?,?,?,?,?,?)"
            
            my_data=(NAME,my_phone,my_address,my_DOR,my_time,my_class,my_age,my_gender,my_pizza,my_MOP)
            my_conn.execute(my_query,my_data)
            my_conn.commit()
            x=my_conn.execute('''select last_insert_rowid()''')
            id=x.fetchone()
            label5.grid() 
            label5.config(fg='green') # foreground color 
            label5.config(bg='white') # background color 
            my_str.set("SUCCESSFUL:") 
            label5.after(3000, lambda: label5.grid_remove() )
            t1.delete('1.0',END)  # reset the text entry box
            
        

        except sqlite3.Error as my_error:
            label5.grid() 
            #return error
            label5.config(fg='red')   # foreground color
            label5.config(bg='yellow') # background color
            print(my_error)
            my_str.set(my_error)        
        
     else:
        label5.grid() 
        label5.config(fg='red')   # foreground color
        label5.config(bg='yellow') # background color
        my_str.set("check inputs.")
        label5.after(3000, lambda: label5.grid_remove() )

my_w.mainloop()
my_conn.close()
```

```python
import tkinter as tk
from pandas import DataFrame
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

data1 = {'PIZZA FLAVOR': ["CHEESY ALOHA","CHESSY PEPERONI","CHEESY MEAT","CHEESY HAM",'CHEESY LOADED','CHEESY VEGGIES','CHEESY BURGER'],
         'TOTAL OF ORDER': [10,5,4,7,12,19,20 ]
        }
df1 = DataFrame(data1,columns=['PIZZA FLAVOR','TOTAL OF ORDER'])

my_w = tk.Tk() 

my_w .geometry("500x650") 
my_w.title("KAMBAL PIZZA!!")
figure1 = plt.Figure(figsize=(50,15), dpi=50)
ax1 = figure1.add_subplot(111)
bar1 = FigureCanvasTkAgg(figure1, my_w )
bar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
df1 = df1[['PIZZA FLAVOR','TOTAL OF ORDER']].groupby('PIZZA FLAVOR').sum()
df1.plot(kind='bar', legend=True, ax=ax1)
ax1.set_title('KABLA PIZZA')

my_w.mainloop()
```
